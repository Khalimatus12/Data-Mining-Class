```python
import numpy as np  
import matplotlib.pyplot as plt  
import pandas as pd  
from apyori import apriori
```

```python
data = pd.read_csv('online_retail.csv')
data.head()
```

```python
data.shape
```

```python
np.sum(data.isnull())
```

```python
#Menghapus missing value pada variabel describe
data['Description'] = data['Description'].str.strip()
data.dropna(axis=0, subset=['InvoiceNo'], inplace=True)
```

```python
#Menghapus pembelian kredit pada variabel Invoiceno dengan awalan c
data['InvoiceNo'] = data['InvoiceNo'].astype('str')
data = data[~data['InvoiceNo'].str.contains('C')]
```

```python
#Menghapus missing value pada describe
data.dropna(axis=0, subset=['Description'], inplace=True)
```

```python
# menghapus missing value pada variabel customer id
data.dropna(axis=0, subset=['CustomerID'], inplace=True)
```

```python
# menghapus missing value pada data country
data.dropna(axis=0, subset=['Country'], inplace=True)
```

```python
np.sum(data.isnull())
```

```python
a = data.groupby('InvoiceNo')
retail = []
for i,j in a:
    retail.append(list(j['Description'].map(str)))
```

```python
data_retail =pd.DataFrame(retail)
```

```python
data_retail.head()
```

```python
data_retail.shape
```

```python
records=[]
for i in range (0, 18536):
    records.append([str(data_retail.values[i,j]) for j in range (0,542)])
```

```python
association_rules=apriori(records, min_support=0.005,
                         min_confidence=0.7,
                         min_lift=3, min_length=2)
association_result=list(association_rules)
```

```python
association_result[0]
```

```python
for item in association_result:

    # first index of the inner list
    # Contains base item and add item
    pair = item[0] 
    items = [x for x in pair]
    print("Rule: " + items[0] + " -> " + items[1])

    #second index of the inner list
    print("Support: " + str(item[1]))

    #third index of the list located at 0th
    #of the third index of the inner list

    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print("=====================================")
```
