## Imputation Missing Value

```python
np.sum(data.isnull()) #number of NA for each variable, jumlah missing value per variabel
```

```python
data.isnull().sum() #sama kayak yang atas
```

```python
data.isnull().sum().sum() #penjumlahan missing value semua variabel
```

```python
#imputasi missing value data numerik
data['A/G'] = data['A/G'].fillna((data['A/G'].mean())) 
data.isnull().any()
```

```python
#imputasi data kategorik dengan modus sesuai kolom masing2
data = data.apply(lambda x:x.fillna(x.value_counts().index[0]))
data.isnull().any()
```

## Encode Labels

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# Encode untuk Sex, Steroid, Anaroxia, Spleen Pabable, Spiders tanpa membuat kolom baru
data['Gender'] = LabelEncoder().fit_transform(data['Gender'])
data.head()
```
